```python
import matplotlib.pyplot as plt
%matplotlib inline
from collections import defaultdict
from datetime import datetime, timedelta

import matplotlib as mpl
mpl.rcParams['figure.figsize'] = (15,10)
```

```python
import pymysql
import pymysql.cursors

def pool():
    # Connect to the database
    return pymysql.connect(
        host='localhost',
        user='andres',
        password='3661071a',
        db='crypto',
        cursorclass=pymysql.cursors.DictCursor)
```

```python
with pool() as cursor:
    x = []
    websites_data = defaultdict(lambda: ([], []))
    query = "SELECT * FROM ticker WHERE coin2='SNT' AND created_at>'2017-07-24' ORDER BY created_at DESC LIMIT 10000"
    cursor.execute(query)
    result = cursor.fetchall()
    for entry in result:
        time_to_insert = entry['created_at']
        if entry['website'] == 'etherdelta':
            time_to_insert + timedelta(hours=1)
        websites_data[entry['website']][0].append(time_to_insert)
        websites_data[entry['website']][1].append(entry['last'])
        
    plots = []
    legends = []
    for website, data in websites_data.items():
        plots.append(plt.plot(data[0], data[1])[0])
        legends.append(website)
    plt.title('SNT last')
    plt.legend(plots, legends)
    
```

```python
with pool() as cursor:
    websites_data = defaultdict(lambda: ([], []))
    query = "SELECT * FROM ticker WHERE coin2='ADX' AND created_at>'2017-07-24'"
    cursor.execute(query)
    result_tickers = cursor.fetchall()
    query = "SELECT * FROM trade WHERE trade_coin='ADX' AND type='sell' AND website_timestamp>'2017-07-24'"
    cursor.execute(query)
    result_trades = cursor.fetchall()
    for entry in result_tickers:
        time_to_insert = entry['created_at']
        #if entry['website'] == 'etherdelta':
        #    time_to_insert + timedelta(hours=1)
        if entry['website'] not in ['etherdelta', 'bittrex']:
            websites_data[entry['website']][0].append(time_to_insert)
            websites_data[entry['website']][1].append(entry['last'])
    ED_trades_x = []
    ED_trades_y = []
    ED_trades_s = []
    for trade in result_trades:
        time_to_insert = trade['website_timestamp'] + timedelta(hours=1)
        ED_trades_x.append(time_to_insert)
        ED_trades_y.append(trade['price'])
        ED_trades_s.append(trade['base_amount'] * 100)
        
    plots = []
    legends = []
    for website, data in websites_data.items():
        plots.append(plt.plot(data[0], data[1])[0])
        legends.append(website)
    plots.append(plt.scatter(ED_trades_x, ED_trades_y, s=ED_trades_s, marker='x', c='r'))
    legends.append('ED trades')
    
    plt.title('ADX last')
    plt.legend(plots, legends)
    
```
